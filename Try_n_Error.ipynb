```python
#SOALAN 1

# Import pustaka
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Menjana matriks data rawak (10 baris x 3 lajur)
data = np.random.randint(1, 100, size=(10, 3))

# Membina DataFrame
df = pd.DataFrame(data, columns=["A", "B", "C"])

# Memaparkan DataFrame
print("Data Rawak:")
print(df)

# Melukis graf garis
plt.figure(figsize=(8, 5))
plt.plot(df["A"], marker='o', label="A")
plt.plot(df["B"], marker='s', label="B")
plt.plot(df["C"], marker='^', label="C")

# Menambah tajuk dan label
plt.title("Graf Data Rawak")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.legend()
plt.grid(True)

# Paparkan graf
plt.show()
```

```python
# Tiada ralat

jumlah = 100
bilangan = 0

if bilangan != 0:
    purata = jumlah / bilangan
    print(purata)
else:
    print("Error: Cannot divide by zero!")
```
